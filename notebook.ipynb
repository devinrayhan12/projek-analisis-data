## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Pada jam berapa terjadi lonjakan peminjaman sepeda tertinggi di hari kerja (working days) dibandingkan akhir pekan (weekend) pada tahun 2012 untuk menentukan waktu pemeliharaan unit yang optimal?
- **Pertanyaan 2:** Bagaimana pengaruh perbedaan kondisi cuaca (weather situation) terhadap jumlah rata-rata peminjaman sepeda harian oleh pengguna registered selama tahun 2011-2012?

### Cleaning Data

#### Fixing ... problem

```python
# 1. Mengubah tipe data dteday menjadi datetime
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])

# 2. Memberikan label deskriptif pada kolom kategorikal agar visualisasi mudah dipahami
# Mapping untuk kondisi cuaca (weathersit)
weather_mapping = {
    1: 'Clear/Few clouds',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Heavy Rain/Ice Pallets'
}
day_df['weather_label'] = day_df['weathersit'].map(weather_mapping)

# Mapping untuk hari kerja agar lebih jelas dalam visualisasi jam
hour_df['workingday_label'] = hour_df['workingday'].map({1: 'Working Day', 0: 'Weekend/Holiday'})

# 3. Verifikasi hasil pembersihan
print("Tipe data dteday setelah cleaning:", day_df["dteday"].dtype)
display(day_df[['dteday', 'weather_label']].head())

# 4. Opsional: Menyimpan data yang sudah bersih untuk dashboard
# day_df.to_csv("dashboard/main_data.csv", index=False)
```

## Exploratory Data Analysis (EDA)

### Explorasi Penyewaan per cuaca

```python

day_df.groupby(by="weathersit").agg({
    "cnt": ["max", "min", "mean", "std"]
})
```

**Insight:**
Berdasarkan hasil statistik di atas, terlihat perbedaan signifikan antara kondisi cuaca. Cuaca 1 (Cerah/Clear) memiliki rata-rata penyewaan tertinggi (sekitar 4000+ per hari) dibandingkan cuaca lainnya. Sebaliknya, pada cuaca 3 (Hujan Ringan/Salju), rata-rata penyewaan turun drastis.

### Penyewa Casual dan Registered (Hari Kerja)

```python
workingday_analysis = day_df.groupby(by="workingday").agg({
    "casual": "mean",
    "registered": "mean",
    "cnt": "mean"
})
print(workingday_analysis)
```

**Insight:** Terdapat pola bahwa pada hari kerja (workingday = 1), jumlah penyewa Registered sangat mendominasi, yang mengindikasikan bahwa sepeda digunakan sebagai alat transportasi utama untuk bekerja atau sekolah. Namun, pada hari libur (workingday = 0), jumlah penyewa Casual meningkat signifikan, menunjukkan bahwa pada hari libur sepeda lebih banyak digunakan untuk kegiatan rekreasi.

### Pola Hari Kerja vs Hari Libur

```python
hourly_peak = hour_df.groupby(by="hr").agg({
    "cnt": ["mean", "max"]
}).sort_values(by=("cnt", "mean"), ascending=False).head(5)
print(hourly_peak)
```

**Insight:**  Analisis data per jam menunjukkan bahwa penyewaan sepeda mencapai puncaknya pada waktu-waktu sibuk (rush hour), yaitu sekitar jam 17:00 dan jam 08:00. Jam 17:00 menjadi waktu paling sibuk secara rata-rata. Informasi ini sangat berguna bagi operasional untuk memastikan ketersediaan sepeda di stasiun-stasiun populer pada jam-jam tersebut.

## Visualization & Explanatory Analysis

```python
# --- EDA untuk Pertanyaan 1 ---
# Menghitung rata-rata penyewa registered berdasarkan kondisi cuaca
weather_analysis = day_df.groupby('weather_label')['registered'].mean().reset_index()

# --- EDA untuk Pertanyaan 2 ---
# Menghitung rata-rata penyewa berdasarkan jam dan tipe hari (hari kerja vs libur)
hourly_pattern = hour_df.groupby(['workingday_label', 'hr'])['cnt'].mean().reset_index()

# Menampilkan hasil untuk memastikan variabel sudah terisi
print("Variabel weather_analysis dan hourly_pattern berhasil didefinisikan!")
display(weather_analysis.head())
```

### Pertanyaan 1:

```python
# EDA
hourly_pattern = hour_df.groupby(['workingday_label', 'hr'])['cnt'].mean().reset_index()

# Visualisasi
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=hourly_pattern,
    x='hr',
    y='cnt',
    hue='workingday_label', # Menggunakan nama kolom yang baru saja dibuat
    marker='o'
)
plt.title('Pola Peminjaman Sepeda per Jam: Hari Kerja vs Hari Libur')
plt.xlabel('Jam (0-23)')
plt.ylabel('Rata-rata Jumlah Peminjam')
plt.xticks(range(0, 24))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Tipe Hari')
plt.show()
```

### Pertanyaan 2:

```python
# EDA 1
weather_analysis = day_df.groupby('weather_label')['registered'].mean().reset_index()

# Visualisasi 1
plt.figure(figsize=(10, 6))
sns.barplot(
    x='weather_label',
    y='registered',
    data=weather_analysis,
    palette='viridis',
    hue='weather_label',
    legend=False
)
plt.title('Rata-rata Peminjaman Sepeda (Registered) Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-rata Peminjam Registered')
plt.show()
```

**Insight:** (Opsional)
- Pada hari kerja, puncak peminjaman terjadi pada jam sibuk seperti jam 08:00  ketika orang berangkat kerja dan 17:00 ketika orang ingin pulang.
- Pada akhir pekan, peminjaman lebih merata dan memuncak di siang hari jam 12:00 - 15:00
